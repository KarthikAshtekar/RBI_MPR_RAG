```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

docs = PyPDFLoader(PDF_PATH).load()
chunks = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=300).split_documents(docs)
print(f"{len(docs)} pages, {len(chunks)} chunks")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectordb = Chroma.from_documents(chunks, embeddings)
print("Vector store ready")
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
retriever = vectordb.as_retriever(search_kwargs={"k": 4})
prompt = ChatPromptTemplate.from_template(
    "Use ONLY this context to answer. If not found, say so.\n\n"
    "Context:\n{context}\n\nQuestion: {question}\n\nAnswer:"
)

def ask(question):
    docs = retriever.invoke(question)
    context = "\n\n".join(d.page_content for d in docs)
    messages = prompt.format_messages(context=context, question=question)
    response = llm.invoke(messages)
    return response.content, docs

print("RAG function ready")
```

```python
answer, docs = ask("What was the repo rate as of April 2025?")
print(answer)
for d in docs:
    print("Page:", d.metadata.get("page") + 1)
```

```python
from sentence_transformers import CrossEncoder
reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")

def retrieve_and_rerank(query, initial_k=15, final_k=4):
    candidates = vectordb.similarity_search(query, k=initial_k)
    pairs = [[query, doc.page_content] for doc in candidates]
    scores = reranker.predict(pairs)
    ranked = sorted(zip(candidates, scores), key=lambda x: x[1], reverse=True)
    return [doc for doc, score in ranked[:final_k]]

print("Re-ranker ready")
```

```python
def retrieval_metrics(retriever, eval_df, k=4):
    hits = 0
    reciprocal_ranks = []
    details = []
    for _, row in eval_df.iterrows():
        docs = retriever.invoke(row["question"])
        pages = [d.metadata.get("page") + 1 for d in docs]
        gt_page = int(row["ground_truth_page"])
        if gt_page in pages:
            hits += 1
            rank = pages.index(gt_page) + 1
            reciprocal_ranks.append(1.0 / rank)
            status = f"HIT (rank {rank})"
        else:
            reciprocal_ranks.append(0.0)
            status = "MISS"
        details.append({"question": row["question"], "ground_truth_page": gt_page,
                         "retrieved_pages": pages, "status": status})
    hit_rate = hits / len(eval_df)
    mrr = sum(reciprocal_ranks) / len(reciprocal_ranks)
    return hit_rate, mrr, pd.DataFrame(details)

hit_rate, mrr, details_df = retrieval_metrics(retriever, eval_df, k=4)
print(f"Hit-rate@4: {hit_rate:.2%}")
print(f"MRR: {mrr:.3f}")
pd.set_option("display.max_colwidth", 60)
details_df
```

```python
def reranked_retrieval_metrics(eval_df, k=4):
    hits = 0
    reciprocal_ranks = []
    for _, row in eval_df.iterrows():
        docs = retrieve_and_rerank(row["question"], initial_k=15, final_k=k)
        pages = [d.metadata.get("page") + 1 for d in docs]
        gt_page = int(row["ground_truth_page"])
        if gt_page in pages:
            hits += 1
            rank = pages.index(gt_page) + 1
            reciprocal_ranks.append(1.0 / rank)
        else:
            reciprocal_ranks.append(0.0)
    hit_rate = hits / len(eval_df)
    mrr = sum(reciprocal_ranks) / len(reciprocal_ranks)
    return hit_rate, mrr

hit_rate_rerank, mrr_rerank = reranked_retrieval_metrics(eval_df, k=4)
print(f"Hit-rate@4 (with re-ranker): {hit_rate_rerank:.2%}")
print(f"MRR (with re-ranker): {mrr_rerank:.3f}")
```

```python
def hybrid_retrieve(query, k=4, rrf_k=60):
    # Get ranked results from both retrievers
    bm25_docs = bm25_retriever.invoke(query)
    dense_docs = vectordb.as_retriever(search_kwargs={"k": 10}).invoke(query)

    # Score each unique chunk using Reciprocal Rank Fusion
    scores = {}
    doc_lookup = {}

    for rank, doc in enumerate(bm25_docs):
        key = doc.page_content  # use content as unique identifier
        scores[key] = scores.get(key, 0) + 1 / (rrf_k + rank + 1)
        doc_lookup[key] = doc

    for rank, doc in enumerate(dense_docs):
        key = doc.page_content
        scores[key] = scores.get(key, 0) + 1 / (rrf_k + rank + 1)
        doc_lookup[key] = doc

    # Sort by combined RRF score, descending
    ranked_keys = sorted(scores.keys(), key=lambda k: scores[k], reverse=True)
    return [doc_lookup[key] for key in ranked_keys[:k]]

print("Custom hybrid RRF retriever ready")
```

```python
def hybrid_retrieve_and_rerank(query, k=4, initial_pool=15, rrf_k=60):
    # Get a wider candidate pool from both retrievers before re-ranking
    bm25_docs = BM25Retriever.from_documents(chunks, k=initial_pool).invoke(query)
    dense_docs = vectordb.as_retriever(search_kwargs={"k": initial_pool}).invoke(query)

    # RRF fusion to build one merged candidate list
    scores = {}
    doc_lookup = {}
    for rank, doc in enumerate(bm25_docs):
        key = doc.page_content
        scores[key] = scores.get(key, 0) + 1 / (rrf_k + rank + 1)
        doc_lookup[key] = doc
    for rank, doc in enumerate(dense_docs):
        key = doc.page_content
        scores[key] = scores.get(key, 0) + 1 / (rrf_k + rank + 1)
        doc_lookup[key] = doc

    ranked_keys = sorted(scores.keys(), key=lambda k: scores[k], reverse=True)
    candidates = [doc_lookup[key] for key in ranked_keys[:initial_pool]]

    # Cross-encoder re-ranks the RRF-fused candidate pool
    pairs = [[query, doc.page_content] for doc in candidates]
    ce_scores = reranker.predict(pairs)
    reranked = sorted(zip(candidates, ce_scores), key=lambda x: x[1], reverse=True)
    return [doc for doc, score in reranked[:k]]

print("Hybrid + Re-ranked retriever ready")
```

```python
def hybrid_reranked_metrics(eval_df, k=4):
    hits = 0
    reciprocal_ranks = []
    for _, row in eval_df.iterrows():
        docs = hybrid_retrieve_and_rerank(row["question"], k=k)
        pages = [d.metadata.get("page") + 1 for d in docs]
        gt_page = int(row["ground_truth_page"])
        if gt_page in pages:
            hits += 1
            rank = pages.index(gt_page) + 1
            reciprocal_ranks.append(1.0 / rank)
        else:
            reciprocal_ranks.append(0.0)
    hit_rate = hits / len(eval_df)
    mrr = sum(reciprocal_ranks) / len(reciprocal_ranks)
    return hit_rate, mrr

hit_rate_hybrid_rerank, mrr_hybrid_rerank = hybrid_reranked_metrics(eval_df, k=4)
print(f"Hit-rate@4 (Hybrid + Re-ranked): {hit_rate_hybrid_rerank:.2%}")
print(f"MRR (Hybrid + Re-ranked): {mrr_hybrid_rerank:.3f}")
```

```python
!pip install -q deepeval

from deepeval.models.base_model import DeepEvalBaseLLM
from deepeval.metrics import GEval, ContextualRelevancyMetric, ContextualRecallMetric
from deepeval.test_case import LLMTestCase, LLMTestCaseParams
import time

class GroqJudge(DeepEvalBaseLLM):
    def __init__(self, model_name="llama-3.1-8b-instant"):
        self.model = ChatGroq(model=model_name, temperature=0)
    def load_model(self):
        return self.model
    def generate(self, prompt: str) -> str:
        return self.model.invoke(prompt).content
    async def a_generate(self, prompt: str) -> str:
        result = await self.model.ainvoke(prompt)
        return result.content
    def get_model_name(self):
        return "Groq-Judge"

judge = GroqJudge()

correctness = GEval(
    name="Correctness",
    criteria="Determine if the actual output is factually consistent with the expected output.",
    evaluation_params=[LLMTestCaseParams.ACTUAL_OUTPUT, LLMTestCaseParams.EXPECTED_OUTPUT],
    model=judge, threshold=0.5,
)
relevancy = ContextualRelevancyMetric(model=judge, threshold=0.5)
recall = ContextualRecallMetric(model=judge, threshold=0.5)
print("Judge and metrics ready")
```

```python
def measure_with_retry(metric, tc, max_retries=5):
    for attempt in range(max_retries):
        try:
            metric.measure(tc)
            return True
        except Exception as e:
            err = str(e).lower()
            if "rate_limit" in err or "429" in err:
                wait = 20 * (attempt + 1)
                print(f"  Rate limited, waiting {wait}s...")
                time.sleep(wait)
            elif "json" in err or "invalid" in err:
                print(f"  Bad JSON from judge, retrying (attempt {attempt+1})...")
                time.sleep(5)
            else:
                raise e
    return False

# Build test cases (uses re-ranked retrieval)
test_cases = []
for _, row in eval_df.iterrows():
    docs = retrieve_and_rerank(row["question"], initial_k=15, final_k=4)
    context = "\n\n".join(d.page_content for d in docs)
    messages = prompt.format_messages(context=context, question=row["question"])
    response = llm.invoke(messages)
    test_cases.append(LLMTestCase(
        input=row["question"],
        actual_output=response.content,
        expected_output=row["expected_answer_summary"],
        retrieval_context=[d.page_content for d in docs],
    ))
    time.sleep(3)

print(f"{len(test_cases)} test cases built")
```

```python
import os
results_list = []
checkpoint_path = "eval_results_checkpoint_rbi.csv"

if os.path.exists(checkpoint_path):
    results_df = pd.read_csv(checkpoint_path)
    results_list = results_df.to_dict("records")
    done_questions = set(results_df["question"])
else:
    done_questions = set()

for i, tc in enumerate(test_cases):
    if tc.input in done_questions:
        continue
    print(f"Scoring test case {i+1}/{len(test_cases)}...")
    measure_with_retry(correctness, tc); time.sleep(8)
    measure_with_retry(relevancy, tc); time.sleep(8)
    measure_with_retry(recall, tc); time.sleep(8)
    results_list.append({
        "question": tc.input,
        "correctness": correctness.score,
        "relevancy": relevancy.score,
        "recall": recall.score,
    })
    pd.DataFrame(results_list).to_csv(checkpoint_path, index=False)

results_df = pd.DataFrame(results_list)
print(results_df)
print()
print("Average Correctness:", results_df["correctness"].mean())
print("Average Contextual Relevancy:", results_df["relevancy"].mean())
print("Average Contextual Recall:", results_df["recall"].mean())
```
